```python
import random
def delete_edge(G, p):
    ebunch = random.sample(G.edges, (int)(len(G.edges)*p))
    G.remove_edges_from(ebunch)
    return G
```

## turan_graph

## planted clique
Ê¨°Êï∞„ÅåÂ§ö„Åô„Åé„Å¶ÂàÜ„Åã„Çâ„Å™„ÅÑ

```python
import random
def add_clique(G,k):
    n = nx.number_of_nodes(G)
    nodes = [i for i in range(n)]
    lis = random.sample(nodes, k)
    for i in range(k):
        for j in range(i+1, k):
            G.add_edge(lis[i], lis[j])
```

```python
# test
m = 4
n = 4
name = "triangular_lattice_graph"
G = nx.triangular_lattice_graph(m, n)
# relabel
labels = {}
for i ,(node) in enumerate(G.nodes):
    labels[node] = i
G = nx.relabel_nodes(G, labels)

add_clique(G, 6)
nx.draw(G)
plt.show()
```

```python
n = 3000
p = 0.5
k =40
name = "fast_gnp_random_graph_planted_clique"
G = nx.fast_gnp_random_graph(n,p)
add_clique(G,k)
nx.write_edgelist(G, f"{save_dir}/{name}_n{n}_p{p}_k{k}.txt", data=False)
```

## add_star to grid

```python
import random
def add_star(G, num, size):
    n = nx.number_of_nodes(G)
    nodes = [i for i in range(n)]
    for i in range(num):
        lis = random.sample(nodes, size)
        nx.add_star(G,lis)
    
```

```python
# test
m = 5
n = 5
name = "triangular_lattice_graph"
G = nx.triangular_lattice_graph(m, n)
# relabel
labels = {}
for i ,(node) in enumerate(G.nodes):
    labels[node] = i
G = nx.relabel_nodes(G, labels)

add_star(G, 0, 6)
nx.draw(G)
plt.show()
```

```python
m = 4
n = 4
name = "grid_2d_graph_add_star"
G = nx.grid_2d_graph(m, n)

# relabel
labels = {}
for i ,(node) in enumerate(G.nodes):
    labels[node] = i
H = nx.relabel_nodes(G, labels)


num = 1
size = 5
add_star(H, num, size)

nx.write_edgelist(H, f"{save_dir}/{name}_m{m}_n{n}_num{num}_size{size}.txt", data=False)
```
