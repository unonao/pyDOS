```python
import networkx as nx
import matplotlib.pyplot as plt
save_dir = "generated_graphs"
```

# Classic

## balanced_tree
Returns the perfectly balanced r-ary tree of height h.

```python
children = 2
hight = 5
G = nx.balanced_tree(r=children, h=hight)
nx.write_edgelist(G, f"{save_dir}/balanced_tree_r{children}_h{hight}.txt", data=False)
```

```python
nx.draw(G)
plt.show()
```

## barbell_graph
Returns the Barbell Graph: two complete graphs connected by a path.
Two identical complete graphs ùêæùëö1 form the left and right bells, and are connected by a path ùëÉùëö2.

```python
Km1 = 5
Pm2 = 3
name = "barbell_graph"
G = nx.barbell_graph(Km1, Pm2)
nx.write_edgelist(G, f"{save_dir}/{name}_Km1{Km1}_Pm2{Pm2}.txt", data=False)
```

```python
nx.draw(G)
plt.show()
```

## circular_ladder_graph
Returns the circular ladder graph ùê∂ùêøùëõ of length n.

ùê∂ùêøùëõ consists of two concentric n-cycles in which each of the n pairs of concentric nodes are joined by a|n edge.

```python
n = 5
name = "circular_ladder_graph"
G = nx.circular_ladder_graph(n)
nx.write_edgelist(G, f"{save_dir}/{name}_n{n}.txt", data=False)

```

```python
nx.draw(G)
plt.show()
```
